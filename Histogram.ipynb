### Write a Python program using OpenCV to perform histogram equalization on the given grayscale image **"parrot.jpg"**.

The program should:

- Import the required libraries (**OpenCV, NumPy, Matplotlib**).
- Read the image **"parrot.jpg"** in **grayscale** format.
- Display the grayscale image and plot its histogram.
- Perform **histogram equalization** on the grayscale image using `cv2.equalizeHist()` to enhance its contrast.
- Using Matplotlib, display the following in a **2 × 2 grid layout**:
  - Original Grayscale Image
  - Histogram of Original Image
  - Enhanced (Equalized) Image
  - Histogram of Enhanced Image


```python

import cv2
import numpy as np
import matplotlib.pyplot as plt


```

```python
# Read the image in grayscale format

image = cv2.imread("Parrot.jpg", 0)
imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Display the grayscale image.
plt.imshow(image, cmap = 'gray')
plt.axis('off')
plt.show()

```

```python
# Plot the histogram of the grayscale image
plt.hist(image.ravel(), 256, [0, 255])
plt.show()

```

```python
# Perform histogram equalization
EqualizeImage = cv2.equalizeHist(image)
plt.imshow(EqualizeImage, cmap = "gray")
plt.axis('off')
plt.show()

```

```python
# Display [1] the Original Image (Gray Image) and its Histogram, and [2] the Enhanced Image and its Histogram using a 2×2 layout in Matplotlib.

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap = 'gray')
plt.title("Original - Gray")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.hist(image.ravel(), bins = 256, range = [0,256])
plt.title("Histogram - Original")

plt.subplot(2, 2, 3)
plt.imshow(EqualizeImage, cmap = 'gray')
plt.title("Equalized - Gray")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.hist(EqualizeImage.ravel(), bins = 256, range = [0,256])
plt.title("Histogram - Equalized")

plt.tight_layout()
plt.show()


```

### Write a Python program using OpenCV to perform color image enhancement through histogram equalization in the HSV color space on the given image **"parrot.jpg"**.

The program should:

- Import the required libraries (**OpenCV, NumPy, Matplotlib**).
- Read the color image **"parrot.jpg"**.
- Plot the **histogram of the original color image** (use the **B, G, and R** channels).
- Convert the image from **BGR to HSV** color space.
- Apply **histogram equalization** to one specific channel to improve brightness and contrast using `cv2.equalizeHist()`.
- Convert the enhanced HSV image back to **BGR** format.
- Using Matplotlib, display the following in a **2 × 2 grid layout**:
  - Original Color Image
  - Histogram of Original Image (BGR channels)
  - Enhanced Color Image
  - Histogram of Enhanced Image (BGR channels)


```python
# Import required libraries

ParrotBGR = cv2.imread("parrot.jpg", 1)
ParrotRGB = cv2.cvtColor(ParrotBGR, cv2.COLOR_BGR2RGB)

plt.imshow(ParrotRGB)
plt.title("Original - Colored Image")
plt.axis('off')
plt.show()

```

```python
# Read the colorgiven parrot.jpg image.
b, g, r = cv2.split(ParrotBGR)

plt.figure(figsize=(10,8))

plt.subplot(2, 2, 1)
plt.hist(b.ravel(), bins=256, range=[0,256], color='blue', alpha=0.5)
plt.title("Blue Channel Histogram")

plt.subplot(2, 2, 2)
plt.hist(g.ravel(), bins=256, range=[0,256], color='green', alpha=0.5)
plt.title("Green Channel Histogram")

plt.subplot(2, 2, 3)
plt.hist(r.ravel(), bins=256, range=[0,256], color='red', alpha=0.5)
plt.title("Red Channel Histogram")

plt.tight_layout()
plt.show()


```

```python
ParrotHSV = cv2.cvtColor(ParrotRGB, cv2.COLOR_RGB2HSV)



```

```python

h, s, v = cv2.split(ParrotHSV)
v = cv2.equalizeHist(v)
hsv_equalized = cv2.merge([h, s, v])



```

```python


Enhanced = cv2.cvtColor(hsv_equalized, cv2.COLOR_HSV2BGR)
EnhancedRGB = cv2.cvtColor(Enhanced, cv2.COLOR_BGR2RGB)

```

```python
plt.imshow(EnhancedRGB)
b2, g2, r2 = cv2.split(Enhanced)
plt.axis('off')
plt.title("Enhanced - Image")
plt.show()



```

```python
plt.subplot(2, 2, 1)
plt.imshow(ParrotRGB)
plt.title("Original Color Image")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.hist(b.ravel(), bins = 256, range = [0,256], color = 'blue', alpha = 0.5)
plt.hist(g.ravel(), bins = 256, range = [0,256], color = 'green', alpha = 0.5)
plt.hist(r.ravel(), bins = 256, range = [0,256], color = 'red', alpha = 0.5)
plt.title("Histogram - Original")

plt.subplot(2, 2, 3)
plt.imshow(EnhancedRGB)
plt.title("Enhanced Color Image")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.hist(b2.ravel(), bins = 256, range = [0,256], color = 'blue', alpha = 0.5)
plt.hist(g2.ravel(), bins = 256, range = [0,256], color = 'green', alpha = 0.5)
plt.hist(r2.ravel(), bins = 256, range = [0,256], color = 'red', alpha = 0.5)
plt.title("Histogram - Enhanced")

plt.tight_layout()
plt.show()




```
